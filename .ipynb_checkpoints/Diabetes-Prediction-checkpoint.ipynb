<h1 style="text-align:Center">Diabetes Prediction</h1>

<h3>Importing dependencies</h3>

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

<h3>Data Collection & Analysis</h3>

```python
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv("diabetes.csv")
```

```python
# printing first five rows of the dataframe
diabetes_dataset.head()
```

```python
# getting the number of rows and columns in this dataframe 
diabetes_dataset.shape
```

```python
# getting the statistical measure of the data
diabetes_dataset.describe()
```

```python
# getting number of diabetic and non-diabetic cases
diabetes_dataset['Outcome'].value_counts()
```

##### 0 ---> Non-diabetic
##### 1 ---> diabetic

```python
# getting mean of both the above cases
diabetes_dataset.groupby('Outcome').mean()
```

```python
# seprating the data and label
X = diabetes_dataset.drop(columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

<h3>Data Standarization</h3>

```python
# standarize our input dataset in the same range
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X = scaler.transform(X)
```

<h3>Train Test Split</h3>

```python
# split data into training set and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

<h3>Training the Model</h3>

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training the support vector machine classifier
classifier.fit(X_train, Y_train)
```

<h3>Model Evaluation</h3>

<h4>Accuracy Score</h4>

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(f"Accuracy score of training dataset is {training_data_accuracy}")
```

```python
# accuracy score on the testing data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(f"Accuracy score of test dataset is {test_data_accuracy}")
```

<h3>Making a Predictive System</h3>

```python
input_data = (10,139,80,0,0,27.1,1.441,57)
```

```python
# convert input data to numpy array
input_data_1 = np.asarray(input_data)
print(input_data)
```

```python
# reshape the array as we are predicting for one instance
input_data_reshape = input_data_1.reshape(1, -1)
print(input_data_reshape)
```

```python
# standardize the data
std_data = scaler.transform(input_data_reshape)
print(std_data)
```

```python
# get the result
prediction = classifier.predict(std_data)
```

```python
# display the result
result = "Diabetic" if prediction[0] == 1 else "Non-Diabetic"
```
